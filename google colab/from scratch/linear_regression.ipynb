```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(-1, 1, 100)
noise = np.random.normal(size = 100,scale=0.5)
y = 3*X + 5 +noise
y = (y - y.mean())/y.var()
epoches = 1000
plt.scatter(X, y)
w_initial = np.random.normal()
b_initial = np.random.normal()
alpha = 0.1
```

```python
def update_w():
  return w_initial - 0.01*derivative_w()
```

```python
def update_b():
  return b_initial - alpha*derivative_b()
```

```python
def derivative_w():
  return np.sum((w_initial * X + b_initial - y)*X) / len(X)
```

```python
def derivative_b():
  return np.sum(w_initial * X + b_initial - y) / len(X)
```

```python
def loss():
  return np.sum((w_initial * X + b_initial - y)**2) / len(X)
```

```python
for j in range(epoches):
  new_b = update_b()
  new_w = update_w()
  w_initial = new_w
  b_initial = new_b
  plt.scatter(j, loss(), color="b")
```

```python
plt.scatter(X, y)
plt.plot(X, X* w_initial + b_initial, color="r")
```
